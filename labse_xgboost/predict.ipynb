```python
import joblib
from sentence_transformers import SentenceTransformer
```

```python
# 1. LaBSE modelini yükle (ilk seferde internetten iner)
labse = SentenceTransformer("sentence-transformers/LaBSE")

# 2. Eğitilmiş XGBoost modelini Drive'dan veya yerelden yükle
model = joblib.load("/content/drive/MyDrive/labse_xgboost_model.pkl")
```

```python
# 3. Tahmin yapmak istediğin yorum
text = "Bu uygulama gerçekten harika!"

# 4. Yorumu embedding'e çevir
embedding = labse.encode([text])  # (1, 768)

# 5. Tahmin yap
prediction = model.predict(embedding)[0]

# 6. Sonucu yazdır
label = "positive" if prediction == 1 else "negative"
print(f"Tahmin sonucu: {label}")
```
