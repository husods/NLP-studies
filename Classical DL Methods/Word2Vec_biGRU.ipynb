```python
from gensim.models import Word2Vec

model_path = "/content/drive/MyDrive/Colab Notebooks/Research Techniques II/Classical DL Methods/azwiki_word2vec.model"
model = Word2Vec.load(model_path)
```

```python
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences

texts = df['text'].str.lower().tolist()
labels = df['labels'].values

tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts)
sequences = tokenizer.texts_to_sequences(texts)

MAX_LEN = 100  # ortalama cümle uzunluğuna göre ayarla
X_seq = pad_sequences(sequences, maxlen=50)
```

```python
import numpy as np
embedding_dim = model.vector_size
word_index = tokenizer.word_index
embedding_matrix = np.zeros((len(word_index) + 1, embedding_dim))

for word, i in word_index.items():
    if word in model.wv:
        embedding_matrix[i] = model.wv[word]
```

```python
from keras.models import Sequential
from keras.layers import Embedding, Bidirectional, GRU, Dense

model = Sequential()
model.add(Embedding(
    input_dim=len(word_index) + 1,
    output_dim=embedding_dim,
    weights=[embedding_matrix],
    input_length=50,  # yukarıda pad_sequences ile belirttiğin MAX_LEN
    trainable=False  # sabit Word2Vec vektörleri
))
model.add(Bidirectional(GRU(64)))
model.add(Dense(1, activation='sigmoid'))  # binary classification

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_seq,
    labels,
    test_size=0.2,
    random_state=42,
    stratify=labels
)

model.fit(X_train, y_train, epochs=5, batch_size=32, validation_data=(X_test, y_test))
```

```python
y_pred = (model.predict(X_test) > 0.5).astype("int32")

from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
