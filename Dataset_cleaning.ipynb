```python
import re

def correct_words_case_insensitive(text, correction_dict):
    def replace(match):
        original = match.group()
        lower = original.lower()
        if lower in correction_dict:
            corrected = correction_dict[lower]
            # Baş harf büyükse, düzeltmeyi de büyük harfli yap
            if original[0].isupper():
                return corrected.capitalize()
            else:
                return corrected
        return original

    pattern = re.compile(r'\b(' + '|'.join(re.escape(k) for k in correction_dict.keys()) + r')\b', flags=re.IGNORECASE)
    return pattern.sub(replace, str(text))
```

```python
import nltk
nltk.download('words')
from nltk.corpus import words
english_vocab = set(words.words())
```

```python
import re

ignore_words = {
    'her', 'sey', 'super', 'ideal', 'bu', 'bele', 'de', 'o', 'zaman', 'cox', 'sade', 'he', 'apar', 'can',
    'seffaf', 'tez', 'tam', 'en', 'son', 'pox', 'aha', 'yer', 'alan', 'ona', 'ala', 'zor', 'bon', 'ba',
    'cetin', 'ne'
}

def is_mostly_english(text):
    tokens = str(text).lower().split()
    if not tokens:
        return False

    # Emojileri ve noktalama işaretlerini silmeden, sadece harf/rakam içerenleri hesaplamaya dahil et
    valid_tokens = [w for w in tokens if re.search(r'[a-zA-Z0-9]', w)]

    if not valid_tokens:
        return False

    english_count = sum(
        1 for word in valid_tokens
        if word in english_vocab and word not in ignore_words
    )

    return english_count / len(valid_tokens) >= 0.51
```

```python
df['truly_english'] = df['text'].apply(is_mostly_english)

# Gerçekten İngilizce yorumları ön izleme
df_eng_check = df[df['truly_english'] == True]

for i, (idx, row) in enumerate(df_eng_check[['text', 'labels']].iterrows(), start=1):
    print(f"{i}. [{row['labels']}] {row['text']}\n")
```

```python
# İngilizce olduğu tahmin edilenleri işaretle
df['is_eng'] = df['text'].apply(is_mostly_english)

# İngilizce olduğu belirlenenleri çıkar
df = df[df['is_eng'] == False].drop(columns=['is_eng']).reset_index(drop=True)
```

```python
import fasttext
import numpy as np

# Modeli yükle
model = fasttext.load_model('lid.176.ftz')

# Dil tahmin fonksiyonu
def debug_language(text):
    cleaned_text = str(text).replace('\n', ' ').strip()
    if not cleaned_text:
        return 'empty'
    labels, probs = model.predict(cleaned_text.lower())
    if not labels or not probs:
        return 'no_label'
    lang_code = labels[0].replace('__label__', '')
    return lang_code

# predicted_lang sütunu yoksa hesapla
if 'predicted_lang' not in df.columns:
    df['predicted_lang'] = df['text'].apply(debug_language)

# Azerice olmayan yorumlar (orijinal kısmın)
df_non_az = df[df['predicted_lang'] != 'az'].copy()
print(f"Non-Azerbaijani yorum sayısı: {len(df_non_az)}")

# Sadece az, tr, en dışındaki yorumları filtrele
excluded_langs = {'az', 'tr', 'en'}
df_other_langs = df[df['predicted_lang'].apply(lambda x: x not in excluded_langs)].copy()
df_other_langs = df_other_langs.reset_index(drop=True)

# Her birini düzgün ve numaralı olarak yazdır
for i, row in df_other_langs.iterrows():
    print(f"{i+1}. [{row['predicted_lang']}] {row['text']}\n")
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_distances
import numpy as np

# LaBSE modeli (çok dilli, yüksek doğruluklu)
model = SentenceTransformer("sentence-transformers/LaBSE")

# Yorumları al
texts = df['text'].astype(str).tolist()

# Embedding hesapla
embeddings = model.encode(texts, show_progress_bar=True, convert_to_numpy=True)

# Ortalama embedding (anlam kümesi merkezi)
mean_vector = np.mean(embeddings, axis=0).reshape(1, -1)

# Kozay uzaklık
distances = cosine_distances(embeddings, mean_vector).reshape(-1)

# Uzaklıkları df'ye ekle
df['embedding_distance'] = distances

# En anlamsız (en uzak) 200 yorumu al
df_outliers = df.sort_values(by='embedding_distance', ascending=False).head(200).reset_index(drop=True)

# Görüntüle
for i, row in df_outliers.iterrows():
    print(f"{i+1}. (Dist: {row['embedding_distance']:.4f}) {row['text']}\n")
```

```python
# Uzaklığı yüksek olanları filtrele (tamamen anlamsız olma ihtimali yüksek)
df_noise = df[df['embedding_distance'] > 1].copy()

# Numaralandırma (okunabilirlik için)
df_noise = df_noise.reset_index(drop=True)
df_noise.insert(0, 'No', df_noise.index + 1)

# Yazdır
for i, row in df_noise.iterrows():
    print(f"{row['No']}. (Dist: {row['embedding_distance']:.4f}) {row['text']}\n")
```

```python
import re

def remove_social_tags(text):
    # @ veya # ile başlayan kelimeleri siler
    return re.sub(r'[@#]\w+', '', str(text))

df['text'] = df['text'].apply(remove_social_tags)
```

```python
import re

def letter_ratio(text):
    text = str(text)
    total_len = len(text)
    if total_len == 0:
        return 0
    letters = re.findall(r'[a-zA-ZəöüğıçşƏÖÜĞİÇŞ]', text)
    return len(letters) / total_len

# Silinecek yorumların maskesi
mask = df['text'].apply(letter_ratio) <= 0.50

# Silme işlemi
df = df[~mask].reset_index(drop=True)
```

```python
import re

def no_alnum(text):
    return not bool(re.search(r'[A-Za-z0-9]', text))

df = df[~df['text'].apply(no_alnum)]
```
